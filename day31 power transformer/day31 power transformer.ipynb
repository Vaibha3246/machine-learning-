```python



import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

import scipy.stats as stats
```

```python
df=pd.read_csv('C:/Users/vaibh/Downloads/concrete_data.csv')

df.head()
```

```python
x=df.iloc[:,0:8]
y=df.iloc[:,-1]
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
from sklearn.model_selection import train_test_split

from sklearn.model_selection import cross_val_score

from sklearn.linear_model import LinearRegression 

from sklearn.preprocessing import PowerTransformer

from sklearn.metrics import r2_score 
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3)
```

```python
# linear regression with out any transsform 


lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_predict=lr.predict(x_test)
```

```python
r2_score(y_test,y_predict)
```

```python
# cross_checking with cross val_score 

lr=LinearRegression()

np.mean(cross_val_score(lr,x,y,scoring='r2'))
```

```python
# ploting the distplot without any transform 
```

```python
for col in x_train.columns:
    
    plt.figure(figsize=(14,5))
    
    plt.subplot(121)
    sns.distplot(x_train[col])
    plt.title(col)
    
    
    plt.subplot(122)
    
    stats.probplot(x_train[col],dist='norm',plot=plt)
    
    plt.title(col)
    
    plt.show()
```

```python
# apply box-transform 

```

```python
pt=PowerTransformer(method='box-cox')

x_train_transform=pt.fit_transform(x_train + .000001)
x_test_transform=pt.fit_transform(x_test + .000001)
```

```python
pd.DataFrame({'col':x_train.columns,'box_cox_lambdas':pt.lambdas_})


#  the value of lambda of each column for beteer normal distribution
```

```python
# linear regression apply on transform
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train_transform , y_train )
# 
```

```python
y_predict2=lr.predict(x_test_transform)
```

```python
r2_score(y_test,y_predict2)
```

```python
# yes accuracy is increased 
```

```python
# using cross cross_val_score 
pt=PowerTransformer(method='box-cox')

x_transform=pt.fit_transform(x + .000001) 

lr=LinearRegression()

np.mean(cross_val_score(lr,x_transform,y, scoring='r2'))
```

```python
x_train_transform=pd.DataFrame(x_train_transform,columns=x_train.columns)
x_train_transform
```

```python
# before and after using box plot 


for col in x_train_transform.columns:
    
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    sns.distplot(x_train[col])
    
    plt.title(col)
    
    plt.subplot(122)
    
    sns.distplot(x_train_transform[col])
    
    plt.title(col)
```

```python
# apply yeo-jhonson
```

```python
# by default it is method=yeo jhonson 

pt=PowerTransformer()

x_train_transform=pt.fit_transform(x_train)
x_test_transform=pt.fit_transform(x_test)

lr=LinearRegression()

lr.fit(x_train_transform,y_train) 

y_predict3=lr.predict(x_test_transform)

print(r2_score(y_test,y_predict3))
```

```python
pd.DataFrame({'col':x_train.columns , 'yeo_jhonson': pt.lambdas_})
```
