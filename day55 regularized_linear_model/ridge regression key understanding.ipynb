```python
alpha=np.linspace(0,30,100)
loss=[]
bias=[]
variance=[]

for i in (alpha):
    reg=Ridge(alpha=i)
    avg_expected_loss,avg_bias,avg_variance=bias_variance_decomp(
        reg,X_train,y_train,X_test,y_test,
        loss='mse',
        random_seed=123)
        
    loss.append(avg_expected_loss)  
    bias.append(avg_bias)
    variance.append(avg_variance)
    
    
    
```
