```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
x=df.drop(columns=['Survived'])
y=df.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=41)
```

```python
imputer1=SimpleImputer(strategy='mean')
imputer2=SimpleImputer(strategy='median')
```

```python
trf=ColumnTransformer([
    
    
    
    ('mean_age',imputer1,[0]),
    ('median_age',imputer2,[1])
],remainder='passthrough')
```

```python
trf
```

```python
trf.fit(x_train)
```

```python
trf.named_transformers_['mean_age'].statistics_
```

```python
trf.named_transformers_['median_age'].statistics_
```

```python
trf.fit_transform(x_train)
trf.fit_transform(x_test)
```
