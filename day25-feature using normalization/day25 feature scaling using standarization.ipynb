```python
df=pd.read_csv('C:/Users/vaibh/Downloads/wine_data.csv',usecols=[0,1,2])
df
```

```python
df.columns=['Class label','Alchool','Malic Acid']
df
```

```python
sns.scatterplot(x='Alchool',y='Malic Acid',data=df,hue=df['Class label'],palette={1:'red',2:'green',3:'pink'})
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(df.drop('Class label', axis=1),df['Class label'],test_size=.3,random_state=0)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler() 

# fit the scaler to train set data , it will learn to the  parameter values

scaler.fit(x_train)

# transform train and test set 

x_train_scaled=scaler.transform(x_train)

x_test_scaled=scaler.transform(x_test)
```

```python
x_train_scaled
```

```python
# this data convert into a numpy array using transform 
# so we convert into a dataframe   
```

```python
x_train_scaled=pd.DataFrame(x_train_scaled,columns=x_train.columns)
x_test_scaled=pd.DataFrame(x_test_scaled, columns=x_test.columns)
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2, figsize=(12,5))

ax1.set_title('before scaling')

ax1.scatter(x='Alchool',y='Malic Acid', data=x_train, c=y_train)

# after scaling  

ax2.set_title('after scaling')

ax2.scatter(x='Alchool',y='Malic Acid', data=x_train_scaled,c=y_train)
```

```python
fig, (ax1,ax2)= plt.subplots(ncols=2, figsize=(12,5))

# before scaling 

# ax1.set_title('before scaling') 
sns.kdeplot(x_train['Alchool'], ax=ax1 ,color='black')

sns.kdeplot(x_train['Malic Acid'], ax=ax1 )


# after scaling 

ax2.set_title('after standard scaling ') 

sns.kdeplot(x_train_scaled['Alchool'],ax=ax2, color='red')
sns.kdeplot(x_train_scaled['Malic Acid'], ax=ax2,color='green' )

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Alcohol Distribution Before Scaling')
sns.kdeplot(x_train['Alchool'], ax=ax1)

# after scaling
ax2.set_title('Alcohol Distribution After Standard Scaling')
sns.kdeplot(x_train_scaled['Alchool'], ax=ax2)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Malic acid Distribution Before Scaling')
sns.kdeplot(x_train['Malic Acid'], ax=ax1)

# after scaling
ax2.set_title('Malic acid Distribution After Standard Scaling')
sns.kdeplot(x_train_scaled['Malic Acid'], ax=ax2)
plt.show()
```
