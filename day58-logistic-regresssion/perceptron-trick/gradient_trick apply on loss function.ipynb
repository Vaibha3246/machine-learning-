```python
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=100,n_features=2,n_informative=1,n_classes=2,n_redundant=0,random_state=41,n_clusters_per_class=1,class_sep=20,hypercube=False)             
```

```python
y.shape
```

```python
import matplotlib.pyplot  as plt
```

```python
plt.figure(figsize=(15,7))
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter')
```

```python
from sklearn.linear_model import LogisticRegression
    
```

```python
lr=LogisticRegression(penalty=None,solver='sag')
```

```python
lr.fit(x,y)
```

```python
x.shape
```

```python
y.shape
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
m=-(lr.coef_[0][0]/lr.coef_[0][1])
b=-(lr.intercept_/lr.coef_[0][1])
```

```python
import numpy as np
```

```python
x_input=np.linspace(-3,3,100)
```

```python
y_input=m * x_input +b
```

```python
plt.figure(figsize=(16,7))
plt.plot(x_input,y_input,c='r')

plt.scatter(x[:,0],x[:,1],c=y,cmap='winter')
plt.ylim(-3,2)
```

```python
def gd(x,y):
    x=np.insert(x,0,1,axis=1)
    weights=np.ones(x.shape[1])
    lr=.6
    for i in range(6000):
        
        y_hat=sigmoid(np.dot(x,weights))
        weights=weights + lr * (np.dot((y-y_hat),x)/x.shape[0])
        
        
    return weights[0],weights[1:]
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
intercept,coef=gd(x,y)
```
