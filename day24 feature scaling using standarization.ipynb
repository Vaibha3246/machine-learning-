```python
df=social.iloc[0:,2:]
df
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(df.drop('Purchased',axis=1),df['Purchased'],test_size=.3,
                                                  random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

standard scaler

```python
from sklearn.preprocessing import StandardScaler 
scaler=StandardScaler()

# fit the scaler to train set , it will learn to the parameters

scaler.fit(x_train)

# transform train and test set 

x_train_scaled=scaler.transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
scaler.mean_
```

```python
x_train_scaled

# it comes in the form of numpy arrays
```

```python
x_train_scaled=pd.DataFrame(x_train_scaled,columns=x_train.columns)
x_test_scaled=pd.DataFrame(x_test_scaled,columns=x_test.columns)
```

```python
x_train_scaled
```

```python
x_test_scaled
```

```python
np.round(x_train_scaled.describe(),1)
```

```python
# effect of scaling 
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))

ax1.scatter(x_train['Age'],x_train['EstimatedSalary'])
ax1.set_title('before scaling')

ax2.scatter(x_train_scaled['Age'],x_train_scaled['EstimatedSalary'],c='purple')

ax2.set_title('after scaling')

plt.show()
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,6))
sns.kdeplot(x_train['Age'],ax=ax1)
sns.kdeplot(x_train['EstimatedSalary'],ax=ax1)
ax1.set_title('before scaling')



sns.kdeplot(x_train_scaled['Age'],ax=ax2)

sns.kdeplot(x_train_scaled['EstimatedSalary'],ax=ax2)
ax2.set_title('after scaling')
plt.show()
```

comparision of distributions 

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,6))

sns.kdeplot(x_train['Age'],ax=ax1)
ax1.set_title('distribution of age before standard scaling')



sns.kdeplot(x_train_scaled['Age'],ax=ax2)
ax2.set_title('distribution of age after  standard scaling')
plt.show()
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2, figsize=(12,6))
# before scaling
ax1.set_title('estimated salary before scaling')
sns.kdeplot(x_train['EstimatedSalary'],ax=ax1)

# after scaling \
 
sns.kdeplot(x_train_scaled['EstimatedSalary'],ax=ax2) 

ax2.set_title('estimated salary after standard scaling')
plt.show()
 
```

why scaling is important 

```python
from sklearn.linear_model import LogisticRegression

lr=LogisticRegression()
lr_scaled=LogisticRegression()

lr.fit(x_train,y_train)

lr.fit(x_train_scaled,y_train)
```

```python
y_predict=lr.predict(x_test)

y_predict_scaled=lr.predict(x_test_scaled)
```

```python
from sklearn.metrics import accuracy_score 
print('actual',accuracy_score(y_test, y_predict))
print('scaled',accuracy_score(y_test,y_predict_scaled))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()

dt_scaled=DecisionTreeClassifier()

dt.fit(x_train,y_train)

dt_scaled.fit(x_train_scaled,y_train)
```

```python
y_predict=dt.predict(x_test)
y_predict_scaled=dt.predict(x_test_scaled)
```

```python
print('actual',accuracy_score(y_test, y_predict))

print('scaled',accuracy_score(y_test, y_predict_scaled))
```
